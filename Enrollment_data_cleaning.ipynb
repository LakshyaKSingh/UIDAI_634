```python
import pandas as pd
```

Starting With Enrollment Data Concatenation

Starting with Enrollment Data

```python
df1=pd.DataFrame(pd.read_csv('api_data_aadhar_enrolment_0_500000.csv'))
df2=pd.DataFrame(pd.read_csv('api_data_aadhar_enrolment_500000_1000000.csv'))
df3=pd.DataFrame(pd.read_csv('api_data_aadhar_enrolment_1000000_1006029.csv'))
```

```python
print("Enrollnent")
print(df1.info())
print(df2.info())
print(df3.info())
```

```python
df_final = pd.concat(
    [df1, df2, df3],
    axis=0,
    ignore_index=True
)
```

```python
df_final.info()
```

```python
df_final.head(30)
```

Date Data Type Correction

```python
df_final['date'] = pd.to_datetime(
    df_final['date'],
    format='%d-%m-%Y',
    errors='raise'
)
```

```python
df_final.info()
```

Derived Measures (Total Enrollment and Year_Month)

```python
df_final['Total_Enrollments'] = (
    df_final['age_0_5'] +
    df_final['age_5_17'] +
    df_final['age_18_greater']
)
```

```python
df_final['year_month'] = (
    df_final['date'].dt.to_period('M').dt.to_timestamp()
)
```

```python
df_final.info()
```

State Cleaning

Mispselled State names and inconsistencies are normalised with basic replacement proceudre

```python
df_final['state'].unique()
```

```python
df_final['state'] = (
    df_final['state']
    .astype(str)
    .str.strip()
    .str.title()
)
```

```python
state_map = {
    # West Bengal mess
    'West Bengal': 'West Bengal',
    'Westbengal': 'West Bengal',
    'West Bangal': 'West Bengal',
    'West Bengal ': 'West Bengal',

    # Odisha / Orissa
    'Orissa': 'Odisha',

    # Andhra
    'Andhra Pradesh': 'Andhra Pradesh',

    # Jammu & Kashmir variants
    'Jammu And Kashmir': 'Jammu and Kashmir',
    'Jammu & Kashmir': 'Jammu and Kashmir',

    # UT mergers / variants
    'Daman And Diu': 'Dadra and Nagar Haveli and Daman and Diu',
    'Daman & Diu': 'Dadra and Nagar Haveli and Daman and Diu',
    'Dadra & Nagar Haveli': 'Dadra and Nagar Haveli and Daman and Diu',
    'Dadra And Nagar Haveli': 'Dadra and Nagar Haveli and Daman and Diu',
    'The Dadra And Nagar Haveli And Daman And Diu':
        'Dadra and Nagar Haveli and Daman and Diu',

    # Andaman variants
    'Andaman & Nicobar Islands': 'Andaman and Nicobar Islands',
    'Dadra And Nagar Haveli And Daman And Diu':
        'Dadra and Nagar Haveli and Daman and Diu',

    'Andaman And Nicobar Islands':
        'Andaman and Nicobar Islands',

    # Puducherry
    'Pondicherry': 'Puducherry',
    '100000': 'Unknown'
}
```

```python
df_final['state'] = df_final['state'].replace(state_map)
```

```python
df_final['state'].unique()
```

```python
df_final['state'].isna().sum()
```

Cleaning Districts

```python
df_final['district'].unique()
```

```python
df_final['district'] = (
    df_final['district']
    .astype(str)
    .str.strip()
    .str.replace(r'\xa0', ' ', regex=True)
    .str.replace(r'\s+', ' ', regex=True)
    .str.replace(r'\*', '', regex=True)
    .str.replace(r'\(.*?\)', '', regex=True)  # remove brackets
    .str.replace(r'[-–—]', ' ', regex=True)
    .str.title()
)
```

```python
district_map = {
    # UP renames
    'Allahabad': 'Prayagraj',
    'Faizabad': 'Ayodhya',
    'Sant Ravidas Nagar Bhadohi': 'Bhadohi',
    'Jyotiba Phule Nagar': 'Amroha',

    # Karnataka
    'Gulbarga': 'Kalaburagi',
    'Bellary': 'Ballari',
    'Mysore': 'Mysuru',
    'Bangalore': 'Bengaluru',
    'Bangalore Rural': 'Bengaluru Rural',
    'Bangalore Urban': 'Bengaluru Urban',

    # Telangana
    'K V Rangareddy': 'Ranga Reddy',
    'K.V.Rangareddy': 'Ranga Reddy',
    'Rangareddi': 'Ranga Reddy',

    # WB
    'Hooghly': 'Hooghly',
    'Hugli': 'Hooghly',
    'Hooghiy': 'Hooghly',
    'Howrah': 'Howrah',
    'Haora': 'Howrah',

    # Odisha
    'Anugul': 'Angul',
    'Sonapur': 'Subarnapur',

    # MP
    'Hoshangabad': 'Narmadapuram',
}
```

```python
df_final['district'] = df_final['district'].replace(district_map)
```

```python
df_final.loc[df_final['district'].isna(), 'district'] = 'Unknown'
```

```python
# How many uniques left?
len(df_final['district'].unique())
```

```python
df_final['district'] = df_final['district'].str.strip()
```

```python
state_district_count = (
    df_final
    .loc[df_final['district'] != 'Unknown']  # exclude junk explicitly
    .groupby('state')['district']
    .nunique()
    .reset_index(name='district_count')
    .sort_values('district_count', ascending=False)
)

state_district_count.head(20)
```

```python
output_path = "enrollment_cleaned_final.csv"

df_final.to_csv(
    output_path,
    index=False,
    encoding="utf-8"
)

print(f"Saved cleaned data to: {output_path}")
print(f"Rows: {len(df_final):,}")
print(f"Columns: {df_final.shape[1]}")
```
