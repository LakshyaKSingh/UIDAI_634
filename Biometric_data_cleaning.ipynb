```python
import pandas as pd
```

Data Loading and Concatenation

```python
files = ["api_data_aadhar_biometric_0_500000.csv", "api_data_aadhar_biometric_500000_1000000.csv", "api_data_aadhar_biometric_1000000_1500000.csv","api_data_aadhar_biometric_1500000_1861108.csv"]
df_final = pd.concat(
    (pd.read_csv(f) for f in files),
    ignore_index=True
)
```

```python
df_final.info()
```

```python
df_final['date'] = pd.to_datetime(
    df_final['date'],
    format='%d-%m-%Y',
    errors='raise'
)
```

```python
df_final.info()
```

```python
df_final['Total Biometric']=(df_final['bio_age_17_']+df_final['bio_age_5_17'])
```

```python
state_map = {
    # Case / spacing / spelling variants
    'WESTBENGAL': 'West Bengal',
    'WEST BENGAL': 'West Bengal',
    'Westbengal': 'West Bengal',
    'West  Bengal': 'West Bengal',
    'West Bangal': 'West Bengal',
    'West bengal': 'West Bengal',
    'west Bengal': 'West Bengal',

    'ODISHA': 'Odisha',
    'odisha': 'Odisha',
    'Orissa': 'Odisha',

    'andhra pradesh': 'Andhra Pradesh',

    'Tamilnadu': 'Tamil Nadu',

    'Chhatisgarh': 'Chhattisgarh',

    'Uttaranchal': 'Uttarakhand',

    # UT / State mergers & renames
    'Pondicherry': 'Puducherry',

    'Daman and Diu': 'Dadra and Nagar Haveli and Daman and Diu',
    'Daman & Diu': 'Dadra and Nagar Haveli and Daman and Diu',
    'Dadra and Nagar Haveli': 'Dadra and Nagar Haveli and Daman and Diu',
    'Dadra & Nagar Haveli': 'Dadra and Nagar Haveli and Daman and Diu',

    'Jammu & Kashmir': 'Jammu and Kashmir',

    'Andaman & Nicobar Islands': 'Andaman and Nicobar Islands',
}
```

```python
df_final['state']=df_final['state'].replace(state_map)
```

```python
df_final['district'] = (
    df_final['district']
    .astype(str)
    .str.strip()
    .str.replace(r'\*+', '', regex=True)
    .str.replace(r'\s+', ' ', regex=True)
    .str.replace('–', '-', regex=False)
    .str.replace('−', '-', regex=False)
)
```
